```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
NewData=pd.read_excel("FinalDataSet_Cleaned.xlsx")
NewData

NewData = NewData.drop('Unnamed: 0', 1)
NewData
```

```python
NewData.describe()
```

```python
plt.scatter(NewData['EffectiveTemperature'],NewData['SurfaceGravity'])
plt.xlabel('Eff. Temp')
plt.ylabel('Surface Gravity')
```

```python
fig = plt.figure(figsize=(8, 6))
plt.scatter(NewData['EffectiveTemperature'],NewData['ApparentMagnitude'])
plt.xlabel('Eff. Temp')
plt.ylabel('Apparent Magnitude')
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot()
ax.text(2,-10,"Hyper Gaints",fontsize = 18)
ax.text(2,-6,"Super Giants",fontsize = 18)
ax.text(3,-4,"Bright Giants",fontsize = 18)
ax.text(2,-1,"Giants",fontsize = 18)
ax.text(5,1,"Sub Gaints",fontsize = 18)
ax.text(1,5,"Main Sequence(Dwarfs)",fontsize = 18)
ax.text(3,7.5,"Sub Dwarfs",fontsize = 18)
ax.text(2,12,"White Dwarfs",fontsize = 18)
ax.text(7,11,"Red Dwarfs",fontsize = 18)
ax.scatter(NewData['SpectralClass'],NewData['AbsoluteMagnitude'],c='g')
plt.xlabel('Spectral Class')
plt.ylabel('Absolute Magnitude')
ax.grid(axis = 'y')
ax.set_xticklabels(['O','B','A','F','G','K','M','L','T','Y'])
ax.invert_yaxis()
plt.show()
```
